```python
# Distribution of Numerical Columns (Age & Fare)

plt.figure(figsize=(12,4))
sns.set_theme(style='darkgrid')

# Age Distribution
plt.subplot(1,2,1)
sns.histplot(data['age'], kde=True, color='lightcoral')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Fare Distribution
plt.subplot(1,2,2)
sns.histplot(data['fare'], kde=True, color='darkblue')
plt.title('Fare Distribution of Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

*Age Insight :*

• Most passengers fall between the age range of 20 to 40 years.  
• The age distribution is slightly right-skewed, indicating fewer elderly passengers.

*Fare Insight :*

• Fare values are right-skewed, with most passengers paying lower fares.  
• A small number of passengers paid significantly higher fares, indicating class differences.


```python
# Countplot of Survival Status (0 = Not Survived, 1 = Survived)

plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=data)
plt.title('Survival Count of Passengers')
plt.xlabel('Survival Status (0 = Not Survived, 1 = Survived)')
plt.ylabel('Number of Passengers')
plt.show()
```

*Survival Insight :*

• The number of passengers who did not survive is higher than those who survived.  
• This shows that survival was not common among the passengers.


```python
# Survival by Gender

plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=data)
plt.title('Survival Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived (0 = No, 1 = Yes)')
plt.show()
```

*Survival by Gender Insight :*

• Female passengers had a higher survival rate compared to male passengers.  
• Gender appears to be an important factor influencing survival outcomes.


```python
# Survival by Passenger Class

plt.figure(figsize=(6,4))
sns.countplot(x='pclass', hue='survived', data=data)
plt.title('Survival Status by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived (0 = No, 1 = Yes)')
plt.show()
```

*Survival by Passenger Class – Insights*

• Passengers traveling in first class had a higher survival rate compared to other classes.  
• Survival chances decreased as passenger class increased.


```python
# Correlation Heatmap (Numerical Relationships)

plt.figure(figsize=(8,5))

numeric_data = data.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

*Interpretation*

• Survival shows correlation with numerical features such as Fare and Pclass.  
• Higher fare values are associated with better survival chances.  
• Correlation analysis helps identify relationships between numerical variables.


```python
# Boxplots for Outlier Detection (Age & Fare)




plt.figure(figsize=(8,6))
sns.set_theme(style='ticks')

# Age Outliers
plt.subplot(1,2,1)
sns.boxplot(y=data['age'], color='green')
plt.title('Age Distribution with Outliers')
plt.ylabel('Age')

# Fare Outliers
plt.subplot(1,2,2)
sns.boxplot(y=data['fare'], color='violet')
plt.title('Fare Distribution with Outliers')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()




```

*Outlier Analysis – Insights*

• The age distribution shows a few outliers, representing very young and elderly passengers.  
• Most passengers fall within a normal age range.  
• Fare values contain several high-value outliers.  
• This indicates a wide variation in ticket prices among passengers.


```python
# Scatter Plot: Age vs Fare (Survival Status)

plt.figure(figsize=(6,4))
sns.scatterplot(x='age', y='fare', hue='survived', data=data)
plt.title('Age vs Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived (0 = No, 1 = Yes)')
plt.show()
```

*Age vs Fare – Insights*

• Passengers who paid higher fares show a higher survival rate.
• There is no clear linear relationship between age and fare.
• Survival appears to be more influenced by fare than age.


```python
# Pair Plot: Relationship between Age, Fare and Survival


sns.set_theme(style='darkgrid')

sns.pairplot(
    data[['age', 'fare', 'survived']],
    hue='survived'
)

plt.suptitle('Pair Plot of Age, Fare and Survival Status', y=1.02)
plt.show()
```

*Pair Plot – Insights*

• The pair plot visualizes relationships between age, fare, and survival status.  
• Passengers with higher fare values tend to have better survival chances.  
• Age alone does not show a strong separation between survived and non-survived passengers.  
• Pair plots help in identifying patterns and relationships between multiple variables.
