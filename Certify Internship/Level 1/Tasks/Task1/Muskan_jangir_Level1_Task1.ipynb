```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading dataset
data = pd.read_csv(r'E:\analytics files\Certify Internship\Level 1\Titanic Dataset.csv')
```

```python
#Overview of dataset
data
```

```python
#lets take a basic info of the dataset
data.info()
```

```python
#check for the missing values
data.isnull().sum()
```

```python
#check missing values visually

plt.figure(figsize=(9,5))
sns.heatmap(data.isnull())
plt.title('Missing Values Before Imputation and Removal')
plt.show()
plt.savefig('missing_values_before.png')
```

```python
#Handling missing values

#Filling missing values in 'age' column with median age
data['age'].fillna(data['age'].median(),inplace=True)
```

```python
#Filling missing values in 'embarked' column with mode
data['embarked'].fillna(data['embarked'].mode()[0],inplace=True)
```

```python
#Filing missing value in 'Fare' column with mean fare
data['fare'].fillna(data['fare'].mean(),inplace=True)
```

```python
#Removing 'cabin' column as it has too many missing values
data.drop('cabin',axis=1,inplace=True)
```

*Dropping Cabin column due to excessive missing values*

```python
#Filling missing values in 'Boat' column with 'None' as it indicates no boat assigned
data['boat'].fillna('None',inplace=True)
```

```python
#Filling missing values in 'Body' column with 0 as it indicates no body found
data['body'].fillna(0,inplace=True)
```

```python
#Filling missing values in 'home.dest' column with 'Unknown' as it indicates unknown destination
data['home.dest'].fillna('Unknown',inplace=True)
```

```python
#again check for missing values
data.isnull().sum()
```

```python
#now check for duplicates

data.duplicated().sum() 
#no duplicates found
#So our data is clean now with no missing values and no duplicates
```

```python
#checking for missing values visually again

plt.figure(figsize=(9,5))
sns.heatmap(data.isnull())
plt.title('Missing Values After Imputation and Removal')
plt.show()
plt.savefig('missing_values_after.png')
```

```python
#save cleaned data to csv

data.to_csv("titanic_clean.csv" , index = False)
```

## Insights

• Missing values were found in the Age, Cabin, Boat, Body, and Home.Dest columns.  
• The Cabin and Body columns contained a high percentage of missing values.  
• Appropriate methods were used to handle missing data, improving dataset quality.  
• After data cleaning, the dataset became more reliable for further analysis.
